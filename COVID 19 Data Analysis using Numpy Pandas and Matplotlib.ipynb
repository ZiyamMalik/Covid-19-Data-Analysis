```python
# Group data by country and sum the totals
country_data = df.groupby("Country")[["Confirmed", "Deaths", "Recovered"]].sum().reset_index()

# Calculate mortality and recovery rates
country_data["Mortality_Rate"] = (country_data["Deaths"] / country_data["Confirmed"]) * 100
country_data["Recovery_Rate"] = (country_data["Recovered"] / country_data["Confirmed"]) * 100

# Display results
country_data[["Country", "Confirmed", "Deaths", "Recovered", "Mortality_Rate", "Recovery_Rate"]].head(10)
```

. Countries Most Affected

The United States, India, and Brazil reported the highest number of confirmed cases globally.

These countries also had a significant death toll, reflecting widespread transmission and healthcare strain.

Smaller nations or island countries recorded relatively fewer cases due to early lockdowns or geographical isolation.

. Trend of Spread Over Months

The number of confirmed cases showed a steep upward trend starting from early 2020, peaking mid-year.

Global daily new cases initially grew exponentially, then began stabilizing as containment measures improved.

The growth curve flattened toward the later stages, indicating improved awareness, testing, and vaccination effects.

. Observations on Death & Recovery Rates

The average global mortality rate was around 2–3%, while the recovery rate was generally 80–90%.

Countries with strong healthcare systems showed higher recovery rates.

Developing nations faced delays in recovery reporting and limited medical infrastructure, which affected their data consistency.

. Key Insights

There’s a strong positive correlation between confirmed and recovered cases — as more people get infected, more recover over time.

Death rates correlate moderately with confirmed cases, suggesting healthcare capacity impacts mortality.

Overall, the dataset reveals how rapid response, testing capacity, and public health measures significantly influence pandemic outcomes.
